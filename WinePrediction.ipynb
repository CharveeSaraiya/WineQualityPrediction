# **Feature Selection**

```python
# Create Classification version of target variable
wine['goodquality'] = [1 if x >= 7 else 0 for x in wine['quality']]# Separate feature variables and target variable
X = wine.drop(['quality','goodquality'], axis = 1)
Y = wine['goodquality']
```

# **Feature Importance**

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

from sklearn.ensemble import ExtraTreesClassifier
classifiern = ExtraTreesClassifier()
classifiern.fit(X,Y)
score = classifiern.feature_importances_
print(score)
```

# **Splitting Dataset**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=7)
```

# **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train,Y_train)
Y_pred = log_reg.predict(X_test)
log_reg_score=log_reg.score(X_train,Y_train)

from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy Score:",accuracy_score(Y_test,Y_pred))
```

```python
confusion_mat = confusion_matrix(Y_test,Y_pred)
print(confusion_mat)
```

# **Using KNN:**

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,Y_train)
y_pred = knn.predict(X_test)
knn_score=knn.score(X_train,Y_train)
from sklearn.metrics import accuracy_score
print("Accuracy Score:",accuracy_score(Y_test,y_pred))
```

# **Using SVC**

```python
from sklearn.svm import SVC
svc = SVC(                                                                                                                                          )
svc.fit(X_train,Y_train)
pred_y = svc.predict(X_test)
svc_score=svc.score(X_train,Y_train)

from sklearn.metrics import accuracy_score
print("Accuracy Score:",accuracy_score(Y_test,pred_y))
```

# **Using Decision Tree:**

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier(criterion='entropy',random_state=7)
DT.fit(X_train,Y_train)
y_pred = DT.predict(X_test)
DT_score=DT.score(X_train,Y_train)
from sklearn.metrics import accuracy_score
print("Accuracy Score:",accuracy_score(Y_test,y_pred))
```

# **Using GaussianNB:**

```python
from sklearn.naive_bayes import GaussianNB
GNB = GaussianNB()
GNB.fit(X_train,Y_train)
y_pred3 = GNB.predict(X_test)
GNB_score=GNB.score(X_train,Y_train)
from sklearn.metrics import accuracy_score
print("Accuracy Score:",accuracy_score(Y_test,y_pred3))
```

# **Using Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(random_state=1)
rf_clf.fit(X_train, Y_train)
y_pred2 = rf_clf.predict(X_test)
rf_clf_score=rf_clf.score(X_train,Y_train)
from sklearn.metrics import accuracy_score
print("Accuracy Score:",accuracy_score(Y_test,y_pred2))
```

# **Using Xgboost**

```python
import xgboost as xgb
XGBR = xgb.XGBClassifier(random_state=1)
XGBR.fit(X_train, Y_train)
y_pred5 = XGBR.predict(X_test)
XGBR_score=XGBR.score(X_train,Y_train)
from sklearn.metrics import accuracy_score
print("Accuracy Score:",accuracy_score(Y_test,y_pred5))
```
